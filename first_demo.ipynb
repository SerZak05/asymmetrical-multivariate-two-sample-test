```python
import numpy as np
import matplotlib.pyplot as plt
from my_tests import *
from experiments import power_mean_experiment, power_var_experiment
from my_statistics import mmd_lin, mmd_gaussian
np.random.seed(42)
```

```python
tests = {
    'assymetrical': reject_asymm_hypo,
    'symmetrical': reject_symm_hypo,
    'mmd2_u (linear kernel)': reject_mmd2_u_gaussian_hypo,
    'mmd2_u (gaussian kernel)': reject_mmd2_u_gaussian_hypo,
    'mmd2_l': reject_mmd2_u_lin_hypo
}

# func_list = (reject_asymm_hypo, reject_symm_hypo, reject_mmd2_u_gaussian_hypo, reject_mmd2_u_gaussian_hypo, reject_mmd2_u_lin_hypo)


def visualize_results(name, tests, x, y_points):
    for i, test in enumerate(tests.keys()):
        plt.plot(x, y_points[i], label=test)
    plt.title(name)
    plt.xlabel("Dimension")
    plt.ylabel("Average power")
    plt.xscale('log')
    plt.legend()
    plt.show()
```

```python
x, y_points = power_mean_experiment(tuple(tests.values()), 50, 100, 100)
```

```python
visualize_results("Normal distributions wit different means", tests, x, y_points)
```

```python
x, y_points = power_var_experiment(tuple(tests.values()), 50, 100, 100)
```

```python
visualize_results("Normal distributions with different variances", tests, x, y_points)
```
