```python
import numpy as np
import matplotlib.pyplot as plt
from my_tests import *
from external_tests import reject_mmd_ext_hypo
from experiments import power_mean_experiment, power_var_experiment, test_significance_level
from my_statistics import mmd_lin, mmd_gaussian
np.random.seed(42)
```

```python
tests = {
    'assymetrical': reject_asymm_hypo, 
    'symmetrical': reject_symm_hypo, # fast
    'mmd2_u (linear kernel)': reject_mmd2_u_lin_hypo, # fast
    # 'mmd2_u (gaussian kernel)': reject_mmd2_u_gaussian_hypo, # 3 min
    'mmd2_l': reject_mmd2_l_hypo # very fast
    # 'mmd_ext (gaussian kernel)': reject_mmd_ext_hypo
}

# func_list = (reject_asymm_hypo, reject_symm_hypo, reject_mmd2_u_gaussian_hypo, reject_mmd2_u_gaussian_hypo, reject_mmd2_u_lin_hypo)


def visualize_results(name, tests, x, y_points):
    for i, test in enumerate(tests.keys()):
        plt.plot(x, y_points[i], label=test)
    plt.title(name)
    plt.xlabel("Dimension")
    plt.ylabel("Average power")
    plt.xscale('log')
    plt.legend()
    plt.ylim(0, 1)
    plt.show()
```

```python
x_mean, y_points_mean = power_mean_experiment(list(tests.values()), 100, 250, 250)
```

```python
visualize_results("Normal distributions with different means", tests, x_mean, y_points_mean)
```

```python
x_var, y_points_var = power_var_experiment(list(tests.values()), 100, 250, 250)
```

```python
visualize_results("Normal distributions with different variances", tests, x_var, y_points_var)
```
